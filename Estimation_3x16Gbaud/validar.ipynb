```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
colores = ['#1f77b4', '#d62728', '#2ca02c', '#000000', '#9467bd', '#8c564b', '#e377c2', '#bcbd22', '#7f7f7f', '#17becf']
```

### Para relu primera capa

```python
hist_relu = np.load("modelos/relu_primer_capa/relu_32_2layers_2classes_eliminado_history.npy", allow_pickle=True).item()
hist_sigmoid = np.load("modelos/relu_primer_capa/sigmoid_32_2layers_2classes_eliminado_history.npy", allow_pickle=True).item()
hist_tanh = np.load("modelos/relu_primer_capa/tanh_32_2layers_2classes_eliminado_history.npy", allow_pickle=True).item()
hist_selu = np.load("modelos/relu_primer_capa/selu_32_2layers_2classes_eliminado_history.npy", allow_pickle=True).item()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(hist_relu['val_accuracy'], label='ReLU', linestyle='dashed', color=colores[0])
plt.plot(hist_tanh['val_accuracy'], label='Tanh', linestyle='dotted', color=colores[1])
plt.plot(hist_selu['val_accuracy'], label='SELU', linestyle='solid', color=colores[2])
plt.plot(hist_sigmoid['val_accuracy'], label='Sigmoid', linestyle='dashdot', color=colores[3])
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.xlim([0, 300])
plt.ylim([0.6, 1])
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.grid(linestyle="--", alpha=0.5, zorder=0)
plt.savefig('relu_activations_32_2layers_2classes_eliminado.svg', format='svg', bbox_inches='tight', pad_inches=0)
plt.show()
```

Prueba hecha para una red neuronal con 32 neuronas, 2 capas ocultas y 2 clases, eliminando 17.6GHz y con relu en la primera capa.

```python
val_relu = np.load("modelos/relu_primer_capa/relu_32_2layers_2classes_eliminado_acc_loss.npy")
val_sigmoid = np.load("modelos/relu_primer_capa/sigmoid_32_2layers_2classes_eliminado_acc_loss.npy")
val_tanh = np.load("modelos/relu_primer_capa/tanh_32_2layers_2classes_eliminado_acc_loss.npy")
val_selu = np.load("modelos/relu_primer_capa/selu_32_2layers_2classes_eliminado_acc_loss.npy")
```

```python
# define los nombres de los modelos
model_names = ['relu', 'sigmoid', 'tanh', 'selu']

# define los valores de accuracy y loss para cada modelo
val_acc = [val_relu[0], val_sigmoid[0], val_tanh[0], val_selu[0]]
val_loss = [val_relu[1], val_sigmoid[1], val_tanh[1], val_selu[1]]

plt.bar(np.arange(len(model_names))-0.15, val_acc, color='green', width=0.3, zorder=3, clip_on=False)
plt.bar(np.arange(len(model_names))+0.15, val_loss, color='blue', width=0.3, zorder=3, clip_on=False)
plt.xlabel('Activation')
plt.ylabel('Accuracy, Loss')
plt.xticks(np.arange(len(model_names)), model_names)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.ylim([0, 1])
plt.legend(['Accuracy', 'Loss'], loc="upper center", ncols=2, bbox_to_anchor=(0.5, 1.1))
plt.grid(linestyle="--", alpha=0.5, zorder=0)
plt.savefig('relu_activations_32_2layers_2classes_eliminado_acc_loss.svg', format='svg', bbox_inches='tight', pad_inches=0)
plt.show()
```
