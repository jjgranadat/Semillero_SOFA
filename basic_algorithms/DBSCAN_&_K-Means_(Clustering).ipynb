# Introduction to K-Means Clustering

```python
import sys
sys.path.append('../my_module')

from sofamethods import bit_error_rate, symbol_error_rate, threshold_density

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import accuracy_score
from sklearn.datasets import make_blobs
```

```python
import warnings
warnings.filterwarnings('ignore')
```

## Input data preparation

```python
X, y = make_blobs(n_samples=500, centers=3, random_state=42, cluster_std=1.6)
ix = np.argsort(y)
X = X[ix]
y = y[ix]

plt.title("Data with 3 clusters and 2 features")
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='rainbow', edgecolors='black')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid()
plt.show()
```

## K-Means & DBSCAN implementation

```python
kmeans = KMeans(n_clusters=3, n_init='auto', init='random').fit(X)
labels = kmeans.labels_

plt.title("K-Means Clustering")
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow', edgecolors='black')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid()
plt.show()

print("Bit Error Rate: %.2f" % (bit_error_rate(y, labels)))
print("Symbol Error Rate: %.2f" % (symbol_error_rate(y, labels)))
print("Accuracy: %.2f%%" % (accuracy_score(y, labels) * 100))
```

```python
dbscan = DBSCAN(eps=1.0, min_samples=5).fit(X)
labels = dbscan.labels_
labels = threshold_density(labels, X, umbral=2.5)

plt.title("DBSCAN Clustering")
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow', edgecolors='black')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid()
plt.show()

print("Bit Error Rate: %.2f" % (bit_error_rate(y, labels)))
print("Symbol Error Rate: %.2f" % (symbol_error_rate(y, labels)))
print("Accuracy: %.2f%%" % (accuracy_score(y, labels) * 100))
```
